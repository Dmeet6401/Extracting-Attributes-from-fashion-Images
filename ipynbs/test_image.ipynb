```python
import tensorflow as tf
import cv2
from PIL import Image
import numpy as np
import collections
import os
```

```python
model=tf.keras.models.load_model('VGG16_150-28-0.67.hdf5')
```

```python
data = []
folder_path = 'test'
for file_name in os.listdir(folder_path):

    # passing the images into the model
    file_path = os.path.join(folder_path, file_name)
    image = np.array(Image.open(file_path))
    IMG_SIZE = (150,150) # for VGG16 model needs this specific img shape
    img = cv2.resize(image,IMG_SIZE)
    img=img/255.0 # scall down image 
    img = np.expand_dims(img, axis=0)
    predict_x= model.predict(img)
    classes_x=np.argmax(predict_x,axis=1)

    data.append((file_name, classes_x))


```

```python
import pandas as pd
# Create a DataFrame from the data list
df33= pd.DataFrame(data, columns=['File', 'Label'])

# Save the DataFrame to a CSV file
df33.to_csv('test_image_predictions_1.csv', index=False)
```
